```python
#Rel plot (Plot of Mean Salary based on Experience in USD)# bivariate
mean_salary = df.groupby(["experience_level"])["salary_in_usd"].mean().sort_values()
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(kind = "line", x = mean_salary.index, y = mean_salary, marker = "o", color = "red")
plt.title("Mean Salary")
plt.show()
```

```python
#Countplot(employee_residence based on work year) #univariate
sns.set_style("darkgrid")
sns.set(rc={'figure.figsize':(20,10)})
sns.countplot(x = "employee_residence", hue = "work_year", data = df)
plt.title("Employee Residence based on work year")
plt.show()
```

```python
# Countplot(Company Location Based on Company Size)
sns.countplot(x="company_location",data=df, hue = "company_size")
plt.title("Company Location vs Company Size")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.barplot(x="remote_ratio", y="salary", hue="work_year", data= df);
```

```python
# pairplot
sns.pairplot(df,hue='job_title')
plt.show()
```

```python
a = sns.PairGrid(df)
a.map_diag(sns.histplot)
a.map_offdiag(sns.scatterplot)
plt.show()
```

```python
# Jointplot (Scatter Plot + Histogram)
sns.set(rc={'figure.figsize':(30,10)})
sns.jointplot(x='salary_currency',y='work_year',data=df,s=300,color='green')
plt.show()
```

```python
#Pie chart(Job_titles)
fig = px.pie(df, values=df['job_title'].value_counts().values, names=df['job_title'].value_counts().index,
             title='<b>Job Titles</b>',
            )
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
# Bar plot (Average Salary in US dollar by experience level)
data = pd.DataFrame(df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=True))
fig = px.bar(data, y='salary_in_usd', x=data.index, text_auto='0.3s',
             hover_data=['salary_in_usd'], color='salary_in_usd',height=400,
             title="<b>Average salary in US dollar by experience level</b>")
fig.show()
```

```python
# boxplot
plt.title("df")
sns.boxplot(df.work_year, df.salary_in_usd, hue=df.experience_level)
```

```python
# Position of the Outlier
print(np.where(df['salary_in_usd']>200000))
```

```python
#Detecting the outliers using IQR and removing them

# Importing
import sklearn

''' Detection '''
# IQR(Inter Quartile Range)
Q1 = np.percentile(df['salary_in_usd'], 25,
				interpolation = 'midpoint')

Q3 = np.percentile(df['salary_in_usd'], 75,
				interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", df.shape)

# Upper bound
upper = np.where(df['salary_in_usd'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df['salary_in_usd'] <= (Q1-1.5*IQR))

''' Removing the Outliers '''
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)

print("New Shape: ", df.shape)
```
